##Set partitions

```python
for fold in np.arange(5):

  expName='integratedPairs_drew+JLM+SCBC_5CV_{0}'.format(fold)
  partitionsFile = workDir + 'models/models_{0}/savedPartitions.pkl'.format(expName)

  dir = workDir + 'dataPartitions/integratedPairs_wSCBC_5CV/integratedPairs_wSCBC_5CV_{0}/'.format(fold)
  trainingData, trainingLabels = dataProcessing.acquire(dir + 'trainingData.tsv', dir + 'trainingLabels.tsv', headings=True)
  testData, testLabels = dataProcessing.acquire(dir + 'testData.tsv', dir + 'testLabels.tsv', headings=True)

  trainingData.drop(columns=['Unnamed: 0'], inplace=True)
  testData.drop(columns=['Unnamed: 0'], inplace=True)

  trainingData.to_csv(dir + 'trainingData.tsv', sep='\t', index=False)
  testData.to_csv(dir + 'trainingData.tsv', sep='\t', index=False)

  _, _, _, _, _, trainingPartitions_sorted = dataProcessing.extractPartitions(trainingData, 'training')
  _, _, _, _, _, testingPartitions_sorted = dataProcessing.extractPartitions(testData, 'testing')
  partitionsConcatenated = np.concatenate((trainingPartitions_sorted, testingPartitions_sorted), axis=0)
  partitionsSorted = np.unique(partitionsConcatenated, axis=0)
  os.makedirs(workDir + 'models/models_{0}'.format(expName), exist_ok=True)
  pickle.dump(partitionsSorted, open(partitionsFile, 'wb'))
```
